```python
import pandas as pd
import matplotlib.pyplot as plt
from fredapi import Fred

fred = Fred(api_key = '*')
```

```python
print("Fetching data from FRED...")
wages = fred.get_series('AHETPI')
cpi = fred.get_series('CPIAUCSL')

df = pd.DataFrame({'Nominal_wage': wages, 'CPI': cpi})

df = df.dropna()
```
