```python
import pandas as pd
import matplotlib.pyplot as plt
from fredapi import Fred

fred = Fred(api_key = '*')
```

```python
print("Fetching data from FRED...")
wages = fred.get_series('AHETPI')
cpi = fred.get_series('CPIAUCSL')

df = pd.DataFrame({'Nominal_wage': wages, 'CPI': cpi})

df = df.dropna()
```

```python
print(df.head(10))
print(df.tail(10))
```

```python
def get_real_wage(Nominal_wage, cpi):
  """
  Take a nominal column and a CPI column
  Returns a new column adjusted for inflation (real dollars)
  """
  cpi_today = wages.iloc[-1]

  real_col = (Nominal_wage / wages)*cpi_today

  return get_real_wage
```

```python
df['Real Wage'] = get_real_wage(df['Nominal_wage'],df['CPI'])
```
